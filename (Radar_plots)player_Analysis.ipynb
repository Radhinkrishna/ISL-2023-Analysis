```python
import pandas as pd
import matplotlib.pyplot as plt
from soccerplots.radar_chart import Radar
```

```python
df=pd.read_csv('lunaanalysis.csv')

q=df['Name ']
q=q[1:]
print(q)
qt=input("Select you player name:")
df['Team']=['kerala blasters','mohan bagan sg','mumbai city fc','Bengaluru fc','Mumbai city fc','Northeast united','fc Goa','fc Goa','Hydrebad fc','Chennain fc','Bengaluru fc','Mohanbagan sg','fc Goa']
```

```python
df2=df[(df['Name ']==('Adrian Luna'))|(df['Name ']==(qt))].reset_index()
params=list(df.columns)
params=params[2:8]

df2
```

```python
ranges=[]
luna=[]
other=[]
for x in params:
  a=min(df2[params][x])
  a=a-(a*0.20)
  b=max(df2[params][x])
  b=b+(b*0.20)
  ranges.append((a,b))

for x in range(len(df2['Name '])):
  if df2['Name '][x]=='Adrian Luna':
    luna=df2.loc[x].values.tolist()
  if df2['Name '][x]==qt:
    other=df2.loc[x].values.tolist()

luna=luna[2:8]
other=other[2:8]
v=[luna,other]
```

```python
title=dict(
    title_name='Adrian Luna',
    title_color='blue',
    subtitle_name='Kerala blasters FC',
    subtitle_color='yellow',
    title_name_2=qt,
    title_color_2='red',
    subtitle_name_2=df2['Team'][1],
    subtitle_color_2='black',
)
credit= '@Godsownfootball'
```

```python
ranges
```

```python
radar=Radar()
fig,ax=radar.plot_radar(ranges=ranges,params=params,values=v,radar_color=('blue','red'),
                        alphs=(0.75,0.6),title=title,endnote=credit,compare=True)
```
